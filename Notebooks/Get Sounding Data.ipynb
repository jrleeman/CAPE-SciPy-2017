```python
from datetime import datetime, timedelta
from metpy.io import get_upper_air_data
import pandas as pd
```

```python
# Set the range of data to retreive (inclusive)
start_date = datetime(1992, 1, 1)
end_date = datetime(1992, 12, 31)

# Times to retreive data (UTC hours)
hours = [0, 12]

# Stations to get data from
stations = ['UIL', 'YLW', 'OTX', 'MFR', 'BOI', 'REV', 'OAK',
            'LKN', 'VEF', 'NKX', 'TFX', 'GGW', 'RIW', 'RAP',
            'SLC', 'GJT', 'DNR', 'ABQ', 'FGZ', 'TUS', 'EPZ',
            'BIS', 'ABR', 'MPX', 'WPL', 'INL', 'YMO', 'GRB',
            'LBF', 'OAX', 'DDC', 'TOP', 'AMA', 'OUN', 'MAF',
            'FWD', 'DRT', 'CRP', 'ADN', 'BRO', 'DVN', 'ILX',
            'SGF', 'LZK', 'SHV', 'LCH', 'JAN', 'LIX', 'BMX',
            'BNA', 'FFC', 'TLH', 'APX', 'DTX', 'ILN', 'WMW',
            'GYX', 'BUF', 'ALB', 'CAR', 'YQI', 'CHH', 'OKX',
            'PIT', 'IAD', 'WAL', 'MHX', 'RNK', 'GSO', 'CHS',
            'JAX', 'TBW', 'MFL', 'KEY']
```

```python
dates = pd.date_range(start = start_date, end = end_date).tolist()
```

```python
def create_date_time_series(ds, n):
    dates = []
    times = []
    for i in range(n):
        dates.append(ds.strftime('%Y-%m-%d'))
        times.append(ds.strftime('%H'))
    return pd.Series(dates), pd.Series(times)
```

```python
def create_station_series(stn, n):
    station = []
    for i in range(n):
        station.append(stn)
    return pd.Series(station)
```

```python
def append_sounding_to_dataframe(ds, df, dt, stn):
    """
    ds : netCDF like object
    df : Pandas dataframe
    dt : datetime stamp
    stn : station
    """
    
    pressure = pd.Series(ds.variables['pressure'][:])
    temperature = pd.Series(ds.variables['temperature'][:])
    dewpoint = pd.Series(ds.variables['dewpoint'][:])
    u_wind = pd.Series(ds.variables['u_wind'][:])
    v_wind = pd.Series(ds.variables['v_wind'][:])
    speed = pd.Series(ds.variables['speed'][:])
    direction = pd.Series(ds.variables['direction'][:])
    
    dates, times = create_date_time_series(dt, len(pressure))
    station = create_station_series(stn, len(pressure))
    
    data_dict = {'Date': dates,
                 'Time': times,
                 'Station': station,
                 'Pressure': pressure,
                 'Temperature': temperature,
                 'Dewpoint': dewpoint,
                 'u wind': u_wind,
                 'v wind': v_wind,
                 'Speed': speed,
                 'Direction': direction}
    
    sounding_data = pd.DataFrame(data_dict, columns=columns)
    df = df.append(sounding_data, ignore_index=True)
    return df
    
```

```python
columns = ['Date', 'Time', 'Station', 'Pressure', 'Temperature', 'Dewpoint', 'u wind', 'v wind', 'Speed', 'Direction']
df = pd.DataFrame(columns=columns)
```

```python
for station in stations:
    for date in dates:
        for time in hours:
            dt = date + timedelta(hours=time)
            print("Getting: ", station, dt)
            try:
                ds = get_upper_air_data(dt, station)
                df = append_sounding_to_dataframe(ds, df, dt, station)
            except:
                print("Error")
```
